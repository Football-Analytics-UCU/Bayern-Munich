```python
from statsbombpy import sb
import pandas as pd
import numpy as np
pd.set_option('display.max_columns', 100)
# pd.set_option('display.max_rows', 100)
pd.set_option("max_colwidth", 200)
```

```python
sb.competitions()
```

```python
def get_num_matches(x):
    try:
        num_ = sb.matches(x['competition_id'], x['season_id']).shape[0]
    except:
        num_ = np.nan
    return num_


df_competitions = sb.competitions()
df_competitions['num_matches'] = df_competitions.apply(lambda x: get_num_matches(x), axis=1)
df_competitions['season_name_id'] = df_competitions[['season_name', 'season_id']].apply(tuple, axis=1)

df_competitions.groupby(['competition_gender', 'competition_international', 'competition_id', 'country_name', 'competition_name'])\
['season_name_id'].apply(list).reset_index().sort_values(['competition_gender', 'country_name'])
```
